```python
import markdown
import os
import time
from datetime import datetime
import shutil
```

```python
def 获得文件的信息(文件名):
    '''获得文件的信息：MD路径,HTML路径,文件名,作者,创建时间,最后修改时间'''
    mdDir = '../MD'
    htmlDir = '../HTML/文章'
    
    MD路径 = mdDir + '/' + 文件名 + ".md"
    HTML路径 = htmlDir + '/' + 文件名 + ".html"
    
    作者 = 'DYL'
    # 获取文件的创建时间
    creation_time = os.path.getctime(MD路径)
    # 获取文件的最后修改时间
    modification_time = os.path.getmtime(MD路径)
    
    # 将时间戳转换为可读格式
    创建时间 = datetime.fromtimestamp(creation_time).strftime('%Y-%m-%d')
    最后修改时间 = datetime.fromtimestamp(modification_time).strftime('%Y-%m-%d')
    return MD路径,HTML路径,文件名,作者,创建时间,最后修改时间
```

```python
def 遍历文件夹(files,Dir,List):
    """遍历并打印每个文件名"""
    for file in files:
        # 构建完整的文件路径
        full_path = os.path.join(Dir, file)
        # 如果是文件（而不是文件夹），则打印
        if os.path.isfile(full_path):
            # 分离文件名和扩展名
            filename, extension = os.path.splitext(file)
            # print(f"文件名: {filename}, 扩展名: {extension}")
            List.append(filename)
    print(Dir,List)
    return List
```

```python
def 检查md文件():
    '''MD文件不存在于HTML中，则将其转换到HTML'''
    mdDir = '../MD'
    htmlDir = '../HTML/文章'
    # 获取指定目录下的所有文件和文件夹名称
    mdfiles = os.listdir(mdDir)
    htmlfiles = os.listdir(htmlDir)
    # 定义两个列表来存放 Markdown和HTML文件名
    mdList = []
    htmlList = []
    # 遍历并打印每个文件名
    遍历文件夹(mdfiles,mdDir,mdList)
    遍历文件夹(htmlfiles,htmlDir,htmlList)
    
    for 文件名 in mdList:
        if 文件名 not in htmlList:
            # print(name)
            MD路径,HTML路径,文件名,作者,创建时间,最后修改时间 = 获得文件的信息(文件名)
            转换md文件(MD路径,HTML路径,文件名,作者,创建时间,最后修改时间)

# 检查md文件()
```

```python

def 将image移动到HTML文件夹():
    # 定义源文件夹路径和目标文件夹路径
    source_folder = '../MD/image'
    destination_folder = '../HTML/文章/image'

    # 检查目标文件夹是否已存在，如果存在则先删除
    if os.path.exists(destination_folder):
        shutil.rmtree(destination_folder)  # 删除目标文件夹及其所有内容

    # 使用shutil.copytree()复制文件夹
    shutil.copytree(source_folder, destination_folder)
    
    # 打印成功的消息
    print(f"文件夹 '{source_folder}' 已成功复制到 '{destination_folder}'")

# 将image移动到HTML文件夹()
```

```python
def 重新生成html(文件名):
    MD路径,HTML路径,文件名,作者,创建时间,最后修改时间 = 获得文件的信息(文件名)
    转换md文件(MD路径,HTML路径,文件名,作者,创建时间,最后修改时间)

def 全部重新生成():
    mdDir = '../MD'
    # 获取指定目录下的所有文件和文件夹名称
    mdfiles = os.listdir(mdDir)
    # 定义两个列表来存放 Markdown文件名
    mdList = []
    # 遍历并打印每个文件名
    遍历文件夹(mdfiles,mdDir,mdList)

    for f in mdList:
        重新生成html(f)
```
